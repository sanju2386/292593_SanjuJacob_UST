```python
import seaborn as sns
import matplotlib.pyplot as plt

# List of numerical columns
numerical_columns = ['Age', 'RestBP', 'Chol', 'Fbs', 'RestECG', 'MaxHR', 'Oldpeak', 'Ca']

# Create a plot for each numerical column
for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='AHD', y=col, data=df)
    
    # Add title and labels
    plt.title(f'Boxplot of {col} by Heart Disease (AHD)', fontsize=14)
    plt.xlabel('Heart Disease (AHD)', fontsize=12)
    plt.ylabel(col, fontsize=12)
    
    # Show the plot
    plt.show()

    # Calculate and print number of outliers for each numerical column
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier threshold
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count the number of outliers
    outliers = df3[(df3[col] < lower_bound) | (df3[col] > upper_bound)]

    print(f'Number of outliers in {col}: {len(outliers)}')
```

```python
import pandas as pd

# Create a list to store the outliers count for each numerical column
outliers_summary = []

# List of numerical columns
numerical_columns = ['Age', 'RestBP', 'Chol', 'Fbs', 'RestECG', 'MaxHR', 'Oldpeak', 'Ca']

# Loop through each numerical column to compute and count outliers
for col in numerical_columns:
    # Calculate IQR for each column
    Q1 = df3[col].quantile(0.25)
    Q3 = df3[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier threshold
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df3[(df3[col] < lower_bound) | (df3[col] > upper_bound)]
    outlier_count = len(outliers)

    # Append the result to the summary list
    outliers_summary.append({
        'Feature': col,
        'Outliers Count': outlier_count
    })

# Convert the summary list into a DataFrame for better visualization
outliers_df = pd.DataFrame(outliers_summary)

# Print the outliers summary
print(outliers_df)
```

```python
# Function to print text in bold
def print_bold(text):
    # ANSI escape code for bold text
    return '\033[1m' + text + '\033[0m'

# Print outliers summary with bold headers
for idx, row in outliers_df.iterrows():
    feature = print_bold(row['Feature'])
    outliers_count = print_bold(str(row['Outliers Count']))
    print(f"{feature}\t\t{outliers_count}")
```

```python
# Insight 1: Age and Heart Disease Prevalence
print("\nInsight 1: Older individuals tend to have a higher prevalence of heart disease.")
bins_age = [20, 30, 40, 50, 60, 70, 80]
labels_age = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins_age, labels=labels_age, right=False)
age_heart_disease = df.groupby('AgeGroup', observed=True)['AHD'].apply(lambda x: (x == 'Yes').mean() * 100)
print("Heart Disease Prevalence by Age Group (%):\n", age_heart_disease)
 
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=age_heart_disease.index, y=age_heart_disease.values)
plt.title('Heart Disease Prevalence by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Percentage with Heart Disease')
plt.show()
```

# insight 2: Chest Pain Type and Heart Disease

```python
print("\nInsight 2: Certain types of chest pain are more strongly associated with heart disease.")
chest_pain_heart_disease = df.groupby('ChestPain')['AHD'].apply(lambda x: (x == 'Yes').mean() * 100).sort_values(ascending=False)
print("Heart Disease Prevalence by Chest Pain Type (%):\n", chest_pain_heart_disease)
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=chest_pain_heart_disease.index, y=chest_pain_heart_disease.values)
plt.title('Heart Disease Prevalence by Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Percentage with Heart Disease')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
